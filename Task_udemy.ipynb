```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
Udemy = pd.read_csv(R"C:\Users\moham\Downloads\archive (6)\UdemyCoursesDataset.csv")
Udemy.head(5)
Udemy.set_index('course_id', inplace= True)
```

```python
Udemy['published_timestamp']=pd.to_datetime(Udemy['published_timestamp'])
Udemy['year'] = Udemy['published_timestamp'].dt.year
Udemy
```

```python
Udemy['month']=Udemy['published_timestamp'].dt.month
Udemy.head(60)
```

```python
plt.figure(figsize=(6,4))
sns.countplot( x='is_paid' ,data= Udemy)
plt.title('Number of people who is paid and not ', {'family':'serif'})
plt.show()
```

```python
my_lables=['All Levels','Beginner Level','Intermediate Level','Expert Level']
my_explode=[0.1,0.1,0.10,0.1]
plt.pie(Udemy['level'].value_counts(),explode=my_explode,shadow=True , labels=my_lables,startangle= 90)
plt.title('percentage of diff types level')
plt.show()
```

```python
Udemy['level'].value_counts()
```

```python
mylabeles = ['True' , 'False']
myexplode =[0.1,0.1 ]
plt.pie(Udemy['is_paid'].value_counts() , explode= myexplode , labels= mylabeles ,startangle= 60 ,shadow= True)
plt.axis('equal')
plt.title('paid vs. free courses')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data= Udemy , y='num_subscribers' ,x='is_paid' )
plt.xlabel('Is Paid')
plt.ylabel('Nums of SUBS')
plt.title('Average Demand')
```

```python
plt.figure(figsize= (20,10))
sns.countplot(data= Udemy , x= 'subject',order= Udemy['subject'].value_counts().index ,palette='Set2')
plt.title('Number of Subject')
plt.show()
```

```python
sns.barplot(data= Udemy , x='subject' ,y='num_subscribers')
```

```python
sns.countplot( y= 'subject' , data= Udemy,order= Udemy['subject'].value_counts().index ,palette='Set2')
plt.title('Course count  per Subject')
plt.xlabel('number of people')
```

```python
plt.figure(figsize=(6,12))
sns.boxplot(x='subject' , y='price', data= Udemy,palette='rocket')
plt.xticks(rotation=45)
plt.title('Price Dist By Subject')
```

```python
Udemy['content_minutes'] = Udemy['content_duration'].str.extract(r'(\d+\.?\d*)').astype(float)
Udemy['content_minutes'] *= Udemy['content_duration'].str.contains('hour').astype(int) * 60 + Udemy['content_duration'].str.contains('min').astype(int) * 1
sns.histplot(Udemy['content_minutes'], bins=30,palette='rocket', kde=True)
plt.ylim(0,1000)
plt.title('Content Duration Distribution (minutes)')
```
